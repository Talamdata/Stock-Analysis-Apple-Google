```python
# Step 1: Import Libraries
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Set plot style
sns.set(style="darkgrid")

# Step 2: Define the stock tickers
apple = yf.Ticker("AAPL")     # Apple Inc.
google = yf.Ticker("GOOGL")   # Alphabet Inc.

# Step 3: Download historical data
start_date = "2020-01-01"
end_date = "2024-12-31"

apple_data = apple.history(start=start_date, end=end_date)
google_data = google.history(start=start_date, end=end_date)

# Step 4: Plot the closing prices
plt.figure(figsize=(12, 6))
plt.plot(apple_data['Cloose'], label="Apple (AAPL)", color="blue")
plt.plot(google_data['Close'], label="Google (GOOGL)", color="red")
plt.title("Stock Price Comparison: Apple vs Google (20202024)", fontsize=16)
plt.xlabel("Date")
plt.ylabel("Closing Price (USD)")
plt.legend()
plt.tight_layout()
plt.show()
```
